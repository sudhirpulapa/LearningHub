```python
import numpy as np
from datetime import datetime

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

def dot_product(a, b):
    result = 0
    for i, j in zip(a, b):
        result = result + i * j
    return result

before = datetime.now()

for i in range(1000000):
    dot_product(a, b)

after = datetime.now()

print("The time taken by traditional Python:", after - before)
```

```python
before = datetime.now()

for i in range(1000000):
    np.dot(a, b) # Numpy library 

after = datetime.now()

print("The time taken by Numpy library:", after - before)
```

| Feature               | List                                   | Array (NumPy)                          |
|-----------------------|----------------------------------------|----------------------------------------|
| **Creation**          | Python built-in, `[1, 2, 3]`            | NumPy array, `np.array([1, 2, 3])`      |
| **Homogeneity**       | Can hold elements of different types   | Homogeneous (same data type)           |
| **Operations**        | Limited, element-wise operations        | Extensive, supports vectorized operations|
| **Memory Efficiency** | Less memory efficient                  | More memory efficient                  |
| **Performance**       | Slower for numerical operations        | Faster for numerical operations        |
| **Functions**         | Limited mathematical functions         | Extensive mathematical functions       |
| **Size Change**       | Dynamic, can change size dynamically   | Fixed size (after creation)            |
| **Extensions**        | Limited functionalities               | Extensive functionalities (NumPy)      |
| **Syntax**            | Simple syntax for basic operations     | More complex syntax for operations     |
| **Dependencies**      | Part of Python standard library         | Requires NumPy library                 |
| **Use Cases**         | General-purpose, diverse use cases     | Scientific computing, data analysis    |

| Feature          | `arange()`                                      | `linspace()`                                   |
|------------------|-------------------------------------------------|-----------------------------------------------|
| **Function Signature** | `numpy.arange([start, ]stop, [step, ]...)`    | `numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)` |
| **Step Value**   | Specifies the step size between values.         | Specifies the total number of values.         |
| **Endpoint Inclusion** | Excludes the stop value by default (can include with `endpoint` parameter). | Includes both start and stop values by default (can be modified with `endpoint` parameter). |
| **Return Type**  | Returns an ndarray with evenly spaced values.  | Returns an ndarray with a specified number of evenly spaced values. |

- E.g. (2, 3, 4):

    - 3 represents the number of rows in every 2D Arrays
    - 4 represents the number of columns in every 2D Arrays
 


```python
np.zeros((5,))
```

```python
np.zeros((5, 2))
```

```python
np.zeros((2, 3, 4), dtype = int)
```

```python
np.zeros((2, 2, 3, 4), dtype = int)
```

**(2, 2, 3, 4):**
- 4D Array means a group of 3D Arrays
- 2: two 3D Arrays
- Every 3D Arrays constitutes 2D Arrays
- Every 2D Arrays constitutes 3 rows and 4 columns
- Total number of elements = 48

| Attribute          | `np.zeros(shape, dtype)`   | `np.empty(shape, dtype)`   |
|---------------------|-----------------------------|-----------------------------|
| Initialization      | Initializes all elements to 0 | Does not initialize values; may contain arbitrary values from memory |
| Memory Allocation   | Allocates and sets memory to zero | Allocates memory without initialization; faster but values are unpredictable |
| Usage               | Used when a default initial value of 0 is required | Used when performance is critical and you intend to fill the array later |
| Data Type           | Can specify data type with `dtype` parameter | Can specify data type with `dtype` parameter |
| Example             | `np.zeros((2, 3), dtype=int)` | `np.empty((2, 3), dtype=float)` |

Note: Both functions take the shape of the array as the first parameter and allow specifying the data type with the `dtype` parameter. The examples provided are illustrative and can be adjusted based on your specific use case.
